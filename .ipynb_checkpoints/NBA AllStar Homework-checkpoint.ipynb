# What is the average weight of all players who played during this time?

```python
import pandas as pd
import numpy as np
```

```python
nbaallstar = pd.read_csv('NBA All Stars 2000-2016 - Sheet1.csv', sep = ',')
nbaallstar
```

# What team is represented the most during this time?

```python

mostteam = nbaallstar.groupby('Team')['Selection Type'].count().reset_index().sort_values('Selection Type', ascending=False)
mostteam
```

# What draft class is represeted the most during this time?

```python
draft = nbaallstar.groupby('NBA Draft Status').count().reset_index().sort_values('Player', ascending=False)
draft
```

# Are foriegn players more prevalent in the All-Star Game during this time?

```python
foreign = nbaallstar.groupby('Nationality').count().reset_index().sort_values('Selection Type', ascending=False)
foreign
```

# How often are the Western All-Stars voted in by fans?

```python
vote = nbaallstar.groupby('Selection Type').count().reset_index().sort_values('Selection Type', ascending=False)
vote
```

# How many times does Steph Curry make the All-Star game during this time period?

```python
curry = nbaallstar[nbaallstar['Team'] == 'Golden State Warriors'].groupby(['Player']).count()
curry
```

# How many Shooting Gaurds have made the All-Star game during this time?

```python
sguard = nbaallstar.groupby('Pos').count().reset_index().sort_values('Selection Type', ascending=False)
sguard
```

# How many Chicago Bulls players have made the All-Star Game during this time period?

```python
bulls = nbaallstar[nbaallstar['Team'] == 'Chicago Bulls'].count()
bulls
```
